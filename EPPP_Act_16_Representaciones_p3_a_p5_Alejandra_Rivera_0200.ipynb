#### Problema 1
* La resta desconocidaEnunciado: Un número al que se le resta siete da como resultado 18.
* ¿Cuál es el número?Paso 1: Análisis y Resolución
* MatemáticaDefinir la variable: Sea $x$ el número buscado.
* Plantear la ecuación: "Un número ($x$) menos ($-$) siete ($7$) es igual ($=$) a $18$".$$x - 7 = 18$$Despejar
* la variable: Para dejar sola a la $x$, el 7 que está restando
* pasa al otro lado de la igualdad sumando (operación inversa).$$x = 18 + 7$$Solución:$$x = 25$$

```python
# Problema 3: Hallar el número original
resultado_final = 18
valor_restado = 7
resultado_final=int(input("Ingrese el resultado final"))
valor_restado=int(input("Ingrese el valor restado"))
# Despeje: x = 18 + 7
numero_buscado = resultado_final + valor_restado

print(f"El número es: {numero_buscado}")
```

* Problema 4
* Operaciones combinadas (Triple y Suma)Enunciado: El triple de un número más cuatro es igual a 31.
* ¿Cuál es ese número?Paso 1: Análisis y Resolución
* MatemáticaDefinir la variable: Sea $x$ el número.Plantear la
* ecuación: "El triple de un número ($3x$) más ($+$) cuatro ($4$) es igual ($=$) a $31$".$$3x + 4 = 31$$Despejar la
* variable:Primero, movemos el 4 que está sumando.
* Pasa restando:$$3x = 31 - 4$$$$3x = 27$$Luego, el 3 que
* multiplica a $x$ pasa dividiendo:$$x = \frac{27}{3}$$Solución:$$x = 9$$Paso 2: Código en Pyt

```python
# Problema 4: El triple de un número más 4
total = 31
suma = 4
multiplo = 3
total=int(input("Ingrese el resultado total"))
suma=int(input("Ingrese el valor suma"))
multiplo=int(input("Ingrese el multiplo"))

# Paso 1: Restar el valor independiente
diferencia = total - suma
# Paso 2: Dividir entre el múltiplo
x = diferencia / multiplo

print(f"El número buscado es: {int(x)}")
```

* Problema 5: Perímetro de un terrenoEnunciado: Un terreno tiene un largo que mide 5 metros más que el ancho.
* Si el perímetro total es de 50m.
* ¿Cuánto mide cada lado?Paso 1: Análisis y Resolución
* MatemáticaDefinir las variables:Ancho = $x$Largo = $x + 5$ (mide 5 metros más que el ancho)Plantear la ecuación: El perímetro de un rectángulo es la suma de sus cuatro lados (2 largos + 2 anchos).$$2(ancho) + 2(largo) = 50$$$$2(x) + 2(x + 5) = 50$$Simplificar y despejar:Distribuimos el 2: $2x + 2x + 10 = 50$Sumamos las $x$: $4x + 10 = 50$Restamos 10: $4x = 50 - 10 \rightarrow 4x = 40$Dividimos entre 4: $x = \frac{40}{4}$Solución:Ancho ($x$): 10 metros.Largo ($x + 5$): 15 metros.

```python
# Problema 5: Dimensiones del terreno
perimetro_total = 50
diferencia_largo = 5
perimetro_total=int(input("Ingrese el perimetro total"))
diferencia_largo=int(input("Ingrese la diferencia de largo"))

# La ecuación simplificada fue: 4x + 10 = 50
# Donde 10 es 2 veces la diferencia del largo (2 * 5)
ancho = (perimetro_total - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"El ancho mide: {int(ancho)} metros")
print(f"El largo mide: {int(largo)} metros")
```
